```python
# Parameters injected by dsl.run_notebook
num_rows = globals().get("num_rows", 5)
prefix = globals().get("prefix", "item")
seed = int(globals().get("seed", 42))

print("[nb:generate_data] Starting notebook")
print(f"[nb:generate_data] Params -> num_rows={num_rows}, prefix={prefix}, seed={seed}")

import os
import random
from datasets import Dataset

# Optional: use seed to make generation deterministic
random.seed(seed)

rows = []
for i in range(int(num_rows)):
    text = f"{prefix}-{i}"
    rows.append({"id": int(i), "text": text, "length": len(text)})

# Shuffle deterministically based on seed (for demo)
random.shuffle(rows)

dataset = Dataset.from_list(rows)

output_dir = "/tmp/kfp_nb_outputs"
os.makedirs(output_dir, exist_ok=True)
artifact_path = os.path.join(output_dir, "dataset")
print(f"[nb:generate_data] Saving dataset with {len(rows)} rows to {artifact_path}")
dataset.save_to_disk(artifact_path)

print("[nb:generate_data] Finished notebook")
```
